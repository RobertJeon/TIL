```python
balance = 10000
def banking(balance:int):
    while True:
        order, money = map(str, input("{입금/출금/파기} {금액}을 입력해주세요 : ").split())
        if money.isdigit():
            money = int(money)
            if order == "출금":
                if balance >= money:
                    balance -= money
                    print(f"{money}원이 출금되었습니다. 현재 잔고는 {balance}입니다.")
                    if balance == 0:
                        print("통장을 파기합니다.")
                        break
                else:
                    print(f"현재 잔고가 부족입니다. {money-balance}가 부족합니다.")
            elif order == "입금":
                balance += money
                print(f"{money}가 입금되었습니다. 현재 잔고는 {balance}입니다.")
            elif order == "파기":
                print(f"현재 잔고는 {balance} 입니다.")
                order_2=input("계속 거래 하시겠습니까? 예:1,아니오:2)")
                if order_2.isdigit:
                    order_2 = int(order_2)
                    if order_2 == 1:
                        continue
                    else:
                        print("통장을 파기합니다.")
                        break
        else:
            print("금액이 숫자가 아닙니다. 다시 입력해주세요.")
banking(10000)
```
