# 블록그룹이란?
- 하나의 파일이 여러 블록에 나눠서 저장되어야 할 수 있음
- 단편화를 최소화 하기 위해
- 파일 단편화(File Fragmentation)란?
    - 메모리나 디스크와 같은 기억 장치의 저장 공간이 여러 개의 조각으로 나뉘는 현상
    - 기억 장치의 저장 공간을 낭비하게 하거나 읽기 및 쓰기 연산의 속도를 늦추는 문제점을 야기
- 내부 단편화 (Internal fragmentation)
    - 섹터의 크기보다 실제 데이터의 크기가 작을 때 발생
    - 섹터 안에서 문제점이 발생
- 외부 단편화 (External fragmentation)
    - 작업을 수행할 때 사용 가능한 저장 공간이 여러 조각으로 나뉘는 현상
    - 하나의 파일이 실제 디스크에서 연속적인 저장 공간에 할당이 되지 못하기 때문에 디스크 읽기 및 쓰기와 같은 연산의 효율이 떨어지게 됨.

### 그룹 디스크립터
- 슈퍼 블록은 파일시스템 전체에 대한 정보를 담고 있다면, 그룹 디스크립터는 블록 그룹에 대한 정보를 가지고 있습니다.
    - 데이터 블록 비트맵의 위치
    - inode 비트맵의 위치
    - inode 테이블의 위치
    - 블록 그룹의 사용가능한 블록 개수
    - 블록 그룹의 사용가능한 inode 개수
    - 디렉토리의 개수
    - 블록 그룹 상태에 대한 Flag
    - 블록 비트앱의 체크섬 값
    - inode 비트맵의 체크섬 값
    - 그룹 디스크립터의 체크섬 값


### 데이터 블록
- 비트맵은 데이터 블록/inode 용도로 지정된 블록들 중 사용 중인 블록과 사용하지 않는 블록을 체크하기 위한 영역입니다.
- 각 블록의 사용여부를 체크하기 위하여 비트맵 블록에서 각 1bit(0 or 1)씩 사용합니다.

### inode 테이블
- inode 테이블은 파일에 대한 정보를 가지고 있는 inode가 모여 있는 테이블입니다.
- inode는 각각 1개의 파일 또는 디렉토리에 대한 정보를 가지고 있습니다.
    - 파일의 종류 및 권한 정보
    - 파일 소유자의 UID
    - 파일의 크기
    - 파일의 마지막 접근시간 정보
    - 파일의 마지막 변경시간 정보
    - 파일의 마지막 수정시간 정보
    - 파일의 삭제 시간 정보
    - 파일의 소유 그룹 정보
    - 파일의 하드링크 개수
    - 파일이 사용하고 있는 블록 개수 정보
    - 파일 속성에 대한 Flag
    - 실제 파일 데이터가 들어있는 데이터 블록의 주소

`mkfs` - 리눅스 파일시스템을 만들어주는 명령어
사용 예시 : `mkfs -t ext4 /dev/sdb1`

`file {/dev/sdb1}` : 파일정보 확인 UUID 확인가능

UUID = Universally Unique identifier
- 장치명이 변경되어도 시스템은 영향없이 UUID라는 것을 통해서 파일시스템을 찾아간다.
- 여기부터는 섹터가 아닌 블럭으로 부른다.

- `blkid`
- `lsblk`
- `mount {/dev/sdb} {/mnt/new1}` : 디렉토리에 마운트 진행
- `systemctl reboot` : 재시작
- `fstab` : 파일시스템 테이블

fstab에 등록하여야 하는 내용
1. 장치이름/UUID
2. 마운트 포인트
3. 파일 시스템 (ext4인지와 같은 여부)
4. 옵션
5. 덤프유무
6. 체크유무

# 논리 볼륨 관리
- 일반적인 스토리지 관리 방식보다 유연한 스토리지 관리 기능을 제공합니다.

### 물리 볼륨 (physical Volume)
- 디스크 전체 또는 일부를 파티션으로 지정하고 그 파티션으로 물리 볼륨을 생성
- 파티션 단위로 생성
- PE(Physical extent)는 볼륨의 블록 단위

### 볼륨 그룹 (Volume Group)
- 1개 이상의 물리 볼륨의 집합으로 구성
- 물리 볼륨 사이즈의 합 = 사용할 수 있는 볼륨 그룹의 총 사이즈
- 볼륨 그룹을 생성할 때 PE의 크기를 지정

### 논리 볼륨 (Logical Volume)
- 사용자가 최종적으로 사용하는 단위
- VG에서 필요한 크기로 할당받아 LV를 생성

< 이점 >
- 디스크의 공간을 확장해야 할 때, 다운타임 발생하지 않는다.
- 여러 개의 디스크 공산을 합쳐서 하나처럼 사용할 수 있게 됩니다.

# RAID란?
- Redundant Array of Independant Disk
- 2개 이상의 디스크를 병렬로 처리하여 성능 및 안정성을 향상 시키는 방식
- 과거에는 주로 용량이 작은 디스크들을 연결하여 디스크 용량을 높이기 위해서 사용
- 요즘은 디스크 성능 향상에서 나아가 디스크 오류나 데이터 손실 등 장애에 대비하기 위한 용도로 사용
    - 한 번 손실되거나 삭제된 데이터를 복구하기에 시간과 노력이 많이 소요되는 만큼, 데이터 손실을 방지하기 위한 일종의 대비책
- 정리
    - 한 개의 디스크에 데이터를 저장하는 방식이 아닌, 데이터 저장의 성능 및 안정성 확보를 위해서 복수의 디스크를 구성하는 방식을 RAID라고 합니다.

- 하드웨어적인 방법
    - 별도의 어댑터 카드나 메인보드 속 내장 칩을 통해 RAID를 구성하는 경우
- 소프트웨어적인 방법
    - 운영체제 안에서 구현되며, 사용자에게 디스크를 하나의 디스크처럼 보이게 한다.

RAID는 여러 개의 디스크를 하나로 묶어 하나의 논리적 디스크로 작동하게 한다.

레이드 종류

1. 레이드 0 (스트라이프 세트)
    - 데이터 분산 처리
        - 복수개의 디스크에 데이터를 분산해 처리
        - 처리 속도가 향상된다는 강점
        - 하나의 디스크가 문제가 발생하면 레이드0으로 구성된 모든 데이터가 유실된다는 단점
    - 최소 드라이브 개수 : 2
2. 레이드 1
    - 데이터 복제
        - 데이터 처리 시, 동일한 디스크에 똑같이 저장하는 방식
        - 데이터를 중복 기록하여 저장하기 때문에, 디스크가 장애 및 오류 발생 시, 다른 디스크를 통해 데이터를 쉽게 복구할 수 있습니다.
            - 가용성 및 안정성이 높다.
        - 중복으로 저장하기 때문에 실질적으로 사용할 수 있는 용량이 절반으로 줄어듭니다.
3. 레이드 0+1
    - 데이터 분산 처리 + 복제
        - 2개 이상의 디스크를 레이드 0 방식으로 구성하여 데이터를 분산처리하고, 각 각의 디스크에 레이드 1 방식으로 디스크를 추가하여 데이터를 복제하는 방식
4. 레이드 3,4
    - 데이터 분산처리 + 에러 체크
        - 레이드 0과 같이 분산처리방식을 사용하며, 추가로 에러 체크 및 수정을 위한 패리티 정보를 별도의 디스크에 따로 저장하는 방식
        - 레이드 3과 레이드 4는 데이터 저장 단위에 따라 구분
            - 레이드 3 = Byte
            - 레이드 4 = Block
        - 패러티 정보를 별도의 디스크에 저장하기 때문에 디스크가 고장 나더라도 패리티정보를 통해 복구가 가능합니다.
        - 패리티 정보를 디스크에 별도 보관하기 때문에 해당 디스크 손실 시, 복구가 어렵다는 취약점
5. 레이드 5
    - 데이터 분산처리 + 에러 체크
        - 패리티 정보를 별도 디스크에 보관 함으로 써 생기는 약점을 보완하기 위해 나온 방식
        - 패리티 정보를 보관하는 디스크 구성 없이, 일정한 로직에 의해 데이터 분산처리 + 에러체크를 구성
        - 성능, 안정성을 모두 고려한 형태로 서버 구축 시 많이 사용하고 있는 레이드 방식

RAID 1 0 & RAID 0 1

- 공통점 : 용량이 같다.
- 차이점 : RAID 1 0 이 더 안전하다.
